```python
import torch
import torchvision.models as models
import json

vgg16 = models.vgg16(pretrained=True)
```

```python
import numpy as np
import os
filename = "../weights/vgg16-weights.bin"
try:
    os.remove(filename)
except:
    print("file did not exisist")
f = open(filename, 'wb')
for k, v in vgg16.state_dict().items():
    print("saving "+k)
    v.require_grad = False
    binary_format = bytearray(v.numpy().tobytes('C'))
    f.write(binary_format)
f.close()
```

```python
def write_to_file(filename, data):
    with open(filename, 'w') as outfile:
        json.dump(data, outfile)
```

```python
dir_path = "./weights/"

for k, v in vgg16.state_dict().items():
    raw_state_dict = {}
    if isinstance(v, torch.Tensor):
        raw_state_dict[k] = (list(v.size()), v.numpy().tolist())
        write_to_file(dir_path+k+'.json', raw_state_dict)
    else:
        print("State parameter type error : {}".format(k))
        #exit(-1)
```

```python
for k, v in vgg16.state_dict().items():
    print(k, " ", v.dtype, " ",v.shape)
```
